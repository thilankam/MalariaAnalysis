```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
care = pd.read_excel("Malaria_yearwise.xlsx", sheet_name = 'Care_1418')
malaria = pd.read_excel("Malaria_yearwise.xlsx", sheet_name = 'Malaria_1418')
ITN = pd.read_excel("Malaria_yearwise.xlsx", sheet_name = 'ITN_1418')
Maldiag = pd.read_excel("Malaria_yearwise.xlsx", sheet_name = 'Maldiag_1418')
```

```python
care_2014 = care[care['TIME_PERIOD'] == 2014]
care_2018 = care[care['TIME_PERIOD'] == 2018]
plt.figure(figsize = (10, 6))
plt.bar([-0.1, 0.9, 1.9, 2.9, 3.9], care_2014['OBS_VALUE'], width = 0.2, label = 'Year 2014')
plt.bar([0.1, 1.1, 2.1, 3.1, 4.1], care_2018['OBS_VALUE'], width = 0.2, label = 'Year 2018')
plt.xticks([0, 1, 2, 3, 4], care_2014['Geographic area'])
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.ylim([0,100])
plt.title("Disrtibution of Careseeking for febrile children\nacross countries for the year 2014 and 2018")
plt.legend()
plt.show()


malaria_2014 = malaria[malaria['TIME_PERIOD'] == 2014]
malaria_2018 = malaria[malaria['TIME_PERIOD'] == 2018]
plt.figure(figsize = (10, 6))
plt.bar([-0.1, 0.9, 1.9, 2.9, 3.9], malaria_2014['OBS_VALUE'], width = 0.2, label = 'Year 2014')
plt.bar([0.1, 1.1, 2.1, 3.1, 4.1], malaria_2018['OBS_VALUE'], width = 0.2, label = 'Year 2018')
plt.xticks([0, 1, 2, 3, 4], malaria_2014['Geographic area'])
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.ylim([0,100])
plt.title("Distribution of Malaria, first line treatment - percentage of febrile children (under age 5) receiving ACT (first line antimalarial drug), among those receiving any antimalarial drugs\nacross countries for the year 2014 and 2018")
plt.legend()
plt.show()

ITN_2014 = ITN[ITN['TIME_PERIOD'] == 2014]
ITN_2018 = ITN[ITN['TIME_PERIOD'] == 2018]
plt.figure(figsize = (10, 6))
plt.bar([-0.1, 0.9, 1.9, 2.9, 3.9], ITN_2014['OBS_VALUE'], width = 0.2, label = 'Year 2014')
plt.bar([0.1, 1.1, 2.1, 3.1, 4.1], ITN_2018['OBS_VALUE'], width = 0.2, label = 'Year 2018')
plt.xticks([0, 1, 2, 3, 4], ITN_2014['Geographic area'])
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.ylim([0,100])
plt.title("Distribution of ITN use by children - percentage of children (under age 5) who slept under an insecticide-treated mosquito net the night prior to the survey\nacross countries for the year 2014 and 2018")
plt.legend()
plt.show()

Maldiag_2014 = Maldiag[Maldiag['TIME_PERIOD'] == 2014]
Maldiag_2018 = Maldiag[Maldiag['TIME_PERIOD'] == 2018]
plt.figure(figsize = (10, 6))
plt.bar([-0.1, 0.9, 1.9, 2.9, 3.9], Maldiag_2014['OBS_VALUE'], width = 0.2, label = 'Year 2014')
plt.bar([0.1, 1.1, 2.1, 3.1, 4.1], Maldiag_2018['OBS_VALUE'], width = 0.2, label = 'Year 2018')
plt.xticks([0, 1, 2, 3, 4], Maldiag_2014['Geographic area'])
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.ylim([0,100])
plt.title("Distribution of Malaria diagnostics - percentage of febrile children (under age 5) who had a finger or heel stick for malaria testing\nacross countries for the year 2014 and 2018")
plt.legend()
plt.show()
```
